```python
#Thêm thư viện cần thiết
import numpy as np
import os
import sqlite3
import cv2
from PIL import Image
#--------------------------------------------------------------------
# NHAP DU LIEU HINH ANH VA DAT TEN KET NOI CO SO DU LIEU
def insertOrUpdate(id, name):
    #kết nối CSDL
    conn =sqlite3.connect("Face.db")
    #Kiểm tra ID
    query = "SELECT * FROM Face WHERE ID="+str(id)
    cursor = conn.execute(query)
    isRecordExist=0
    for row in cursor:
        isRecordExist=1
    if isRecordExist==0:
         query="INSERT INTO Face(ID, Name) VALUES("+str(id)+",'"+str(name)+"')"     
    else:
          query="UPDATE Face SET Name='"+str(name)+" WHERE ID="+str(id)
    conn.execute(query)
    conn.commit()
    conn.close()

# khai báo 1 lớp nhận diện khuôn mặt lưu vào 1 biến
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")#lọc khuôn mặt của opencv

cap = cv2.VideoCapture(0)# mở camera. .
id = input("Enter user id:" )
name = input("Enter name: ")
insertOrUpdate(id, name)
sample_number = 0
while True:
     # dịch chuyển cái ô hình vuông cho vào mặt mình
    ret, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  #Convert ảnh sang màu xám
    faces = face_cascade.detectMultiScale(gray, 1.3, 5) # Sử dụng phương thức detectMultiScale để phát hiện khuôn mặt trong bức ảnh xám
    for (x,y,w,h) in faces:
        sample_number += 1

        if not os.path.exists("data_face"):
            os.makedirs("data_face")
        #lưu khuôn mặt đã chụp vào thư mục data_face dưới dạng .jpg
        cv2.imwrite("data_face/User."+str(id)+"."+str(sample_number)+".jpg",  img[y:y+h,x:x+w])
        # Vẽ một hình tứ giác xung quanh những khuôn mặt phát hiện được. Vẽ trên ảnh màu.
        cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
        
    cv2.imshow("img", img)
    cv2.waitKey(1);
    if(sample_number>150):
        cap.release()
        cv2.destroyAllWindows()
        break;

```

```python
#training
recognizer = cv2.face.LBPHFaceRecognizer_create() #sử dụng công cụ nhận dạng khuôn mặt LBPHF nằm trong thư viện OPENCV
path = 'data_face'

def getImagesWithID(path):  #lấy hình ảnh từ data
    imagePaths=[os.path.join(path, f) for f in os.listdir(path)]
    faces=[]
    IDs=[]
    for imagePath in imagePaths: #Convert ảnh và add vào mảng faces cùng với ID
        faceImg = Image.open(imagePath).convert('L')
        faceNp = np.array(faceImg, 'uint8')
        ID=int(os.path.split(imagePath)[-1].split('.')[1])# tách ID từ tên file ảnh
        faces.append(faceNp)
        IDs.append(ID)
        cv2.imshow('training', faceNp)  # Hiển thị kết quả ra màn hình
        cv2.waitKey(10)
    return np.array(IDs), faces

Ids, faces = getImagesWithID(path)
recognizer.train(faces, Ids)

if not os.path.exists('trainer'):
    os.makedirs('trainer')

recognizer.save('huanluyen/huanluyen.yml')#Train xong thì lưu thành 1 file huấn luyện
cv2.destroyAllWindows()
```

```python
#Nhận diện
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read("huanluyen/huanluyen.yml")

def getProfile(Id): #Kết nối CSDL và lấy thông tin
    conn=sqlite3.connect("Face.db")
    query="SELECT * FROM Face WHERE ID="+str(Id)
    cursor=conn.execute(query)
    profile=None
    for row in cursor:
        profile=row
    conn.close()
    return profile
cap = cv2.VideoCapture(0)#Mở camera
font = cv2.FONT_HERSHEY_COMPLEX
while True:
    ret, img = cap.read() 
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #convert ảnh
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:  # Hiển thị thông tin User bằng text

        cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 2)# Vẽ một hình tứ giác xung quanh những khuôn mặt phát hiện được                                                               
        roi_gray = gray[y:y+h, x:x+w]                      # Vẽ trên ảnh màu.
        roi_color = img[y:y+h, x:x+w]

        nbr_predicted, conf = recognizer.predict(gray[y:y+h, x:x+w])#công cụ nhận dạng .predict() sẽ lấy tham số là một 
        if conf < 50:                                                #phần của hình ảnh được chụp của khuôn mặt cần phân 
            profile=getProfile(nbr_predicted)                        #tích và sẽ trả về chủ sở hữu của khuôn mặt đó
            if profile != None:
                cv2.putText(img, ""+str(profile[1]), (x+10, y), font, 1, (0,255,0), 1);
        else:
            cv2.putText(img, "Unknown", (x, y + h + 30), font, 0.4, (0, 0, 255), 1);


    cv2.imshow("img", img)# Hiển thị kết quả ra màn hình
    if(cv2.waitKey(1) == ord("b")):# nhấn phím b để thoát chương trình
        break
cap.release()
cv2.destroyAllWindows()
```
